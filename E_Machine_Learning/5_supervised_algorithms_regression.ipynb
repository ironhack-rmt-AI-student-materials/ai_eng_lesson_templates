
# Supervised Learning Algorithms - Regression

In this notebook, we give a **high-level overview** of the most common regression algorithms — organized by family.





## 1. Linear Models

- **Linear Regression** — fits a straight line by minimizing squared errors. The simplest baseline for regression.
- **Polynomial Regression** — extends Linear Regression with polynomial features (e.g. $x^2$, $x_1 x_2$) to capture curved relationships.
- **Ridge (L2)** — adds a penalty on the sum of squared weights to shrink coefficients and reduce overfitting.
- **Lasso (L1)** — adds a penalty on the sum of absolute weights; can shrink coefficients all the way to zero, acting as automatic feature selection.
- **ElasticNet** — combines Ridge and Lasso penalties; useful when there are many correlated features.

## 2. Distance-Based Models

- **KNN Regressor** — predicts by averaging the target values of the K nearest training points. Requires feature scaling; slow on large datasets.

## 3. Tree-Based Models

- **Decision Tree Regressor** — splits data into regions recursively and predicts the mean target value in each region. Highly interpretable but prone to overfitting; `max_depth` is the key parameter.

## 4. Ensemble Methods

- **Random Forest Regressor** *(Bagging)* — builds many decorrelated decision trees on random data/feature subsets and averages their predictions. Strong general-purpose baseline.
- **XGBoost / LightGBM / CatBoost** *(Boosting)* — trains trees sequentially, each correcting the residual errors of the previous. Often the top performer on tabular data.

## 5. Other Algorithms

- **Support Vector Regression (SVR)** — fits the data within a tolerance margin (ε-tube); uses the kernel trick for non-linear patterns. Works well on small datasets.
- **Neural Networks / Deep Learning** — stacked layers of neurons that learn complex representations. Best suited for unstructured data (images, text); tree-based models usually win on tabular data.

## Summary

| # | Model | Family | Core idea | When to use |
|---|---|---|---|---|
| 1 | Linear Regression | Linear | Minimize squared errors with a straight line | Linear relationships, simple baseline |
| 2 | Polynomial Regression | Linear | Add polynomial features to capture curves | Curved but smooth relationships |
| 3 | Ridge / Lasso / ElasticNet | Linear | Penalize large weights to reduce overfitting | Many features, risk of overfitting |
| 4 | KNN Regressor | Distance-based | Average the K nearest neighbors' values | Small datasets, local patterns |
| 5 | Decision Tree Regressor | Tree-based | Split data into regions, predict region mean | Interpretable baseline, non-linear patterns |
| 6 | Random Forest Regressor | Ensemble (Bagging) | Average many decorrelated trees | Strong general-purpose baseline |
| 7 | XGBoost / LightGBM / CatBoost | Ensemble (Boosting) | Sequentially correct residual errors | Best performance on tabular data |
| 8 | Stacking | Ensemble (Stacking) | Meta-model learns to combine base models | Squeezing out extra performance |
| 9 | SVR | Other | Fit within an ε-margin using kernels | Small datasets, non-linear patterns |
| 10 | Neural Networks | Other | Layered representations of the data | Unstructured data, very large datasets |
