# Python Error Handling

In Python, errors generally fall into two main categories: Syntax Errors, which occur when the code violates the language's grammar rules and fails to parse, and Exceptions, which are errors detected during execution (like ZeroDivisionError or TypeError). 

In this notebook, we'll explore some common types of errors, and how you can handle them gracefully using a try-except block.

<br>



<br>






## 1. Types of Errors

<br>

Types of Errors in Python:

- Syntax Errors
    - Code breaks Python’s grammar rules.
    - Caught before execution.
- Runtime Errors (Exceptions)
    - Happen while the program is running.

<br>

## 2. Syntax Errors

Syntax errors are mistakes in the code's structure or grammar that prevent it from running. Python detects these before execution.

<br>



<br>
<br>

````python
#
# Syntax Errors - Example 1
#

employees = ["Alice", "Bob", "Charlie"

for name in employees:
    print(name)
````

## 3. Runtime Errors (Exceptions)

Runtime errors are errors that occur during program execution, often due to unexpected conditions or invalid operations.



<br>
<br>


````python
# Example 1:
# 
# TypeError → Wrong type used

a = 10
# b = input("Please provide a number")
b = "20"

# print(type(b))

result = a + b

result
````

````python
# Example 2:
# 
# ValueError → Correct type, wrong value


b = "hello"
b = float(b)

b
````

````python
# Example 3:
# 
# KeyError → Missing dictionary key

students = ["Alice", "Bob", "Charlie"]

e = students[99]

print(e)
````

````python
# Example 4:
# 
# ZeroDivisionError

total = 10 / 0
````

````python
#
# Example: accessing keys of a dictionary
#


user_details = {
    "name": "Alice", 
    "age": 30,
}

age = user_details["age"]

# email = user_details["email"] # KeyError

email = user_details.get("email")

print(age)
print(email)
````

## 4. Error Handling

Errors are inevitable in programming, but how you handle them defines the robustness of your code. In this section, we’ll explore Python’s try-except blocks—a powerful way to catch and manage exceptions gracefully.

<br>

Syntax:

```py
try:
    # risky code
except SomeError:
    # what to do if it fails
```


<br>

````python
#
# Example of try except
#


numbers = [10, 20, "30", "pizza", 40]

total = 0

for n in numbers:
    try:
        total += float(n)
    except ValueError:
        print(f"Skipping invalid value: {n}... {type(n)}")

print(total)
````

<br>

Why Error Handling Matters
- Prevents crashes
- Makes programs robust
- Lets you control failure behavior
