# Python - OOP


Object-Oriented Programming (OOP) is a programming paradigm that organizes code using classes and objects. It helps structure programs by bundling related properties (attributes) and behaviors (methods) into objects.

Key Concepts:
- Class – A blueprint for creating objects.
- Object – An instance of a class.
- Method – A function in OOP
- Encapsulation – Restricting direct access to data and modifying it only through methods.
- Inheritance – Creating a new class from an existing class.
- Polymorphism – Using a single interface for different data types.
- Abstraction – Hiding complex implementation details and exposing only necessary parts.



```python
# using dictionaries (objects) to store info...

user_1 = {
    "name": "Alice",
    "age": 30,
}

user_2 = {
    "name": "Bob",
    "age": 40,
}
```

```python
#
# Creating a class User
#

class User:
    def say_hello(self):
        print("hello")
    def say_goodbye(self):
        print("goodbye")


u1 = User() # we create an instance of the class User
u1.say_hello()
u1.say_goodbye()

u2 = User()
u2.say_hello()
u2.say_goodbye()

```

```python
#
# Creating a class User +  instance attributes
#

class User:
    def __init__(self, user_name):
        self.user_name = user_name
    def say_hello(self):
        print(f"hello my name is {self.user_name}")
    def say_goodbye(self):
        print("goodbye")


u1 = User("Alice")
u1.say_hello()

u2 = User("Bob")
u2.say_hello()
```

## Inheritance

Inheritance is a core OOP concept where a child class (subclass) inherits attributes and methods from a parent class (superclass). 

This promotes code reusability and establishes a hierarchical relationship.


```python
# parent class
class Animal:
    def __init__(self, species, town):
        self.species = species
        self.town = town
    def say_hello(self):
        print(f"hello, I am a {self.species} and I live in {self.town}")

# child class (inherits from Animal)
class Dog(Animal):
    def __init__(self, dog_name, town):
        super().__init__("dog", town)
        self.dog_name = dog_name
    def bark(self):
        print(f"woof woof, my name is {self.dog_name}")

ant_1 = Animal("ant", "Lisbon")
ant_1.say_hello()

freddy = Dog("Freddy", "Madrid")
freddy.say_hello()
freddy.bark()
```
