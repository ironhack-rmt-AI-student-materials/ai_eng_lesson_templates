## Displaying data

Common options to display data:
1. `print()` is used to display text or data in the console.
2. `display()` is typically used in Jupyter notebooks to render objects in a more formatted or rich display.
3. The last expression in a cell is automatically displayed as output


```python
b = 30

print(b)    # option 1
display(b)  # option 2
b           # option 3
```

## Variables

A variable is like a labeled box where you can store information (like a number, word, or list) in your program.

What we use them for:
- **Storing data**: Variables hold values (e.g., numbers, text, lists) so you can work with them later.
- **Reusability**: Instead of repeating the same value, you reference the variable name.
- **Flexibility**: You can change the value of a variable as your program runs.

<br>


```python
amount = 3.14
user_name = 'alice'

total = amount * 2

print(amount)
print(total)
```

```python
# Dynamic typing is allowed in Python, but generally not considered good practice in production code. 
# Changing a variable's type dynamically can make code harder to read, understand, and maintain.


x = 10
x = 'twenty'

print(x)
```

## Variable names in Python + good practices

```python
# Variables are case-sensitive

amount = 10
Amount = 20

print(amount)
```

```python
# Convention: for variables and functions, we always use snake_case

totalScore = 100 # avoid (in Python) 🚫
total_score = 100 # good (that's the common convention in Python) ✅
```

Python variable naming rules:
1. Can only contain letters (`a–z`, `A–Z`), digits (`0-9`), or underscores `_`.
2. Cannot start with a number
3. Cannot be a Python keyword (if, for, def, class, etc.).
4. Case-sensitive (score ≠ Score).

```python
# 
# Valid names
# 

level = 2 
player1 = "Alice" 
player2 = "Bob" 
is_active = True
total_score = 100
```

```python
# 
# Invalid names
# 
# 1value = 10           # SyntaxError
# total$ = 20           # SyntaxError
# is-logged-in = False  # SyntaxError

```
