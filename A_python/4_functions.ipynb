## Default Parameters




````python
# By default, all parameters must be passed when calling a function

def say_hello(user_name, fav_drink):
    print(f"hello {user_name}, would you like to drink some {fav_drink}")


say_hello("Alice", "coffee")
# say_hello("Bob") # TypeError: say_hello() missing 1 required positional argument: 'fav_drink'

````

````python
# With default parameters

def say_hello(user_name, fav_drink="water"):
    print(f"hello {user_name}, would you like to drink some {fav_drink}")


say_hello("Alice", "coffee")
say_hello("Bob")
````

 Note: 
- Parameters with default values must come after parameters without defaults.
- In other words: Once you define a default parameter, all following parameter must also have defaults.

<br>

### ⚠️ Warning: for default arguments, do not use mutable objects (list, dict, set) 

Avoid using mutable objects (like lists, dicts, or sets) as default parameter values. Default values are evaluated once, when the function is defined — not each time it's called. So a mutable default gets shared and modified across all calls.

These data types should not be used as default parameters (they're mutable objects):
- list
- dict
- set
- bytearray



````python
# "The mutable default argument trap"...
# ...using mutable objects as default parameters can lead to unexpected results

def add_item(item, items=[]):
    items.append(item)
    return items

print(add_item("apple"))   # ['apple']
print(add_item("banana"))  # ['apple', 'banana']  <- unexpected!

#
# Problem: 
# The second call should arguably start fresh, 
# but it reuses the same list from the first call.
#
````

````python
# Solution: use None as the default, and create the mutable object inside the function.

def add_item(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items

print(add_item("apple"))   # ['apple']
print(add_item("banana"))  # ['banana']  <- correct
````

📌 Rule of thumb for default arguments:
- **Never use mutable objects (`[]`, `{}`, `set()`) as default argument values.**
- Instead, use `None` and initialise inside the function body.

    ```python
    # Avoid
    def my_function(data=[]):
        # ...
        # ...

    # Instead, you can do the following
    def my_function(data=None):
        if data is None:
            data = []
        # ...
        # ...
    ```

## Positional Arguments & Keyword Arguments


Positional Arguments:
- Positional arguments are passed to a function based on their position in the function call. The order of the arguments matters,

Keyword Arguments:
- Keyword arguments are passed by specifying the parameter names along with their values in the function call. This allows you to pass arguments out of order or omit some arguments if their default values are defined in the function.

````python
#
# Example 1
#

def greet_guest(guest_name, fav_drink):
    print(f"hello {guest_name}, would you like to drink some {fav_drink}")


# unexpected...
greet_guest("coffee", "alice") # hello coffee, would you like to drink some alice

# solution 1 (with positional arguments - we need to make sure the order is correct):
greet_guest("alice", "coffee")

# solution 2 (with keyword arguments):
greet_guest(fav_drink="coffee", guest_name="alice")
````

Rule of thumb:
- Use positional arguments (solution 1 style) when meaning is obvious and there are few parameters.
- Use keyword arguments (solution 2 style) when clarity matters or there’s any chance of confusion.

````python
#
# Example 2 (default parameters + keyword arguments)
#

def greet_student(student_name, fav_drink="coffee", fav_snack="cake"):
    print(f"hi {student_name}, would you like to drink some {fav_drink} with {fav_snack}")


greet_student("Alice")
greet_student("Bob", "tea")
greet_student("Charlie", "cookies") # unexpected!
greet_student("Charlie", fav_snack="cookies") # yeah, that works :)
````

## Return

- A return is a statement used inside a function to send a result back to the place where the function was called. 
- It allows a function to produce an output that can be stored, used, or passed to other parts of a program.
- Once a function reaches a return statement, it finishes executing and exits.



````python
#
# Example:
#

def calc_sum(a, b):
    total = a + b
    return total
    

result = calc_sum(5, 3)

print(result)
````

````python
#
# Note: a function without a return, will return None
#

def say_hello():
    print("hello world")

r1 = say_hello()

print(r1) # None
````

````python
#
# A function can have multiple return statements, but only one will be executed each time the function runs.
# 
# IMPORTANT: as soon as a return is executed, it will finish the function execution
#

def check_even_odd(amount):
    if amount % 2 == 0:
        return "Even"
    else:
        return "Odd"

r1 = check_even_odd(3)
r2 = check_even_odd(4)

print(r1)
print(r2)
````

## Scope

Scope refers to the region where a variable is accessible.


````python
def do_something():
    amount = 20     # local variable
    print(amount)


do_something()

# print(amount) # NameError: name 'amount' is not defined
````

Some ideas:

- Variables defined inside a function have local scope.
- They are accessible only within the function they are defined in.



#### Scope: "LEGB" rules

- `Local (L)` – Variables defined inside a function; accessible only within that function.
- `Enclosing (E)` – Variables defined in an outer function; accessible within that function and its nested functions.
- `Global (G)` – Variables defined at the module/script level; accessible throughout the module, including from functions.
- `Built-in (B)` – Names provided by Python itself (e.g., print, len); accessible everywhere in Python code.


> The LEGB rule is a kind of name lookup procedure, which determines the order in which Python looks up names. For example, if you reference a given name, Python will look that name up sequentially in the local, enclosing, global, and built-in scope. If the name exists, you get the first occurrence of it. Otherwise, you get an error.


````python
z = 10 # Global scope (will be available thoughout the module/script)

def outer():
    y = 20 # Enclosing scope (will be available within this function and enclosing functions)

    def inner():
        x = 30 # Local Scope
        print(x, y, z) # Accessing local, enclosing, and global
    inner()

outer()

# print(x) # error: from the parent scope, you don't have access to variables in a child scope
````

Main ideas:
- Child scopes: from a child scope, you can access variables in the parent scope.
- Parent scopes: from a parent scope, you cannot access variables defined in a child scope.
- Note: blocks like if, for, and while do not create new scopes in Python — variables defined inside them are accessible in the enclosing function or module scope.


Or, if you prefer:
- "What happens inside of a function stays inside of a function"
- Variables created inside a function are local to that function and aren’t accessible outside unless returned or declared global


<br>


#### The `global` keyword

In Python, global lets a function modify a variable that lives in the module (global) scope.



````python
# Example 1 (without global):

count = 0

def increment():
    count = 5   # without "global", this would create a new variable in the local scope
    print(count)

increment()
print(count) 
````

````python
# Example 2 (with global):

count = 0

def increment():
    global count
    count = 5   # with "global", this will modify the variable in the global scope
    print(count)

increment()
print(count) 
````

Recommendation: 
- Avoid global when possible — it makes code harder to reason about. 
- Instead, prefer using function arguments and return values

## Lambda functions


- A lambda function in Python is a small, anonymous function — meaning it doesn’t need a name. 
- It’s useful when you want a quick, simple function without formally defining one with def.

<br>

Syntax:
```
lambda parameters: expression
```

- `parameters` → input(s) the function takes
- `expression` → what the function returns (it must be a single expression)


````python
#
# Example 1
#

calc_square = lambda x : x * x

print(calc_square(5)) 


# Equivalent to:
#
# def calc_square(x):
#     return x * x
#
````

````python
#
# Example 2
#

calc_total = lambda num1, num2 : num1 + num2

print(calc_total(10, 5))

# Equivalent to:
#
# def calc_total(num1, num2):
#     return num1 + num2
#
````

````python
#
# Example 3 
#

is_even = lambda n : "Even" if n % 2 == 0 else "Odd"

print(is_even(12))
print(is_even(15))


# Equivalent to:
#
# def is_even(n):
#     if n % 2 == 0:
#         return "Even"
#     else:
#         return "Odd"
#
````
