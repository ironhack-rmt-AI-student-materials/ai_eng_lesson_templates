## Operators

### Arithmetic operators: `+`, `-`, `*`, `/`, `//`, `%`, `**`


```python
a = 10
b = 20

print(a * b)
```

```python
x = 10
y = 3

print(x // y)
```

```python
t = 5
t ** 3     # 5 * 5 * 5
```


<br>


### Assignment operators: `=`, `+=`, `-=`, `*=`, `/=`

```python
total = 10

# total = a + 5

total += 5

total
```

<br>

### Relational operators (aka comparison operators) (`==`. `!=`, `<`, `>`, `<=`, `>=`)



```python

amount_A = 50
amount_B = 50

if amount_A >= amount_B:
    print("A is bigger or equals")
else:
    print("it's not")

```

```python

student_A = "Alice"
student_B = "Bob"

if student_A == student_B:
    print("they're the same")
else:
    print("they're different")
```


<br>

### Logical operators (`and`, `or`, `not`)

```python
#
# and
#

day = 'monday'
is_hungry = True

if day == 'friday' and is_hungry is True:
    print('Pizza please')
else:
    print('Salad please!')


```

```python
#
# or
#

age = 42
is_working = True

if age < 18 or is_working is True:
    print('Water please')
else:
    print('Beer please')


#
# Note, this is actually much more common:
# 
# if age < 18 or is_working:
#   ...
#
```

```python
#
# not
#

age = 42
is_working = False

if age > 18 and not is_working:
    print('Beer please')
else:
    print('Water please')
```

##  "==" vs. "is"


`==` compares values
→ Use `==` when you want to know if two objects contain the same data.

```python
a = [1, 2, 3]
b = [1, 2, 3]

if a == b:
    print("they have the same values")
```

`is` compares identity → Use is when you want to know if two variables refer to the exact same object in memory.

```python
a = [1, 2, 3]
b = [1, 2, 3]

if a is b:
    print("they're not the same object in memory")
```

The most common use is checking against None:


```python
model = None

if model is None:
    print("No model loaded")
else:
    print("Model loaded!")


```

Rule of thumb:
- ✅ Use == to compare values.
- ✅ Use is when checking for None.
- ❌ Don't use is to compare numbers, strings, or lists—you'll often get surprising results because is is about identity, not equality.
