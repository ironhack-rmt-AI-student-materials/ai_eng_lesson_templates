## List Comprehension

A Python list comprehension is a concise way to create a new list by applying an expression to each item in an existing iterable.

Examples of iterables: lists, ranges, strings, dictionaries...

<br>


Syntax:

```py
[expression for item in iterable]
[expression for item in iterable if condition]
```

- `expression` → what you want each element of the new list to be
- `item` → variable representing each element in the iterable
- `iterable` → the list, range, or any iterable you loop over
- `if condition` (optional) → filter elements

<br>



Image source: https://cdn-media-0.freecodecamp.org/size/w2000/2021/07/list-comprehension-1.png


````python
#
# Example 1: 
# 
# Goal:
# - Create a list with the square numbers from 0 to 4
# - [0, 1, 4, 9, 16]
#

squares_A = [x * x for x in range(5)]
print(squares_A)


#
# Let's do the same, but without a list comprehension...
#

squares_B = []

for x in range(5):
    squares_B.append(x * x)

print(squares_B)
````

````python
#
# Example 2 (with a condition): 
# 
# Goal:
# - Create a list with the even numbers from 0 to 9
# - [0, 2, 4, 6, 8]
#

nums_list = [n for n in range(10) if n % 2 == 0]

nums_list
````

# Dictionary Comprehension

Dictionary Comprehension in Python is a concise way to create dictionaries from iterables.

Examples of iterables: lists, ranges, strings, dictionaries...

<br>

Syntax:

```py

{key_expr: value_expr for item in iterable if condition}

```

- `key_expr` → expression for dictionary keys
- `value_expr` → expression for dictionary values
- `iterable` → any iterable to loop over
- `if condition` → optional filter

<br>




Image source: https://buggyprogrammer.com/wp-content/uploads/2022/02/Comprehenssion.png 

````python
# 
# Example: Create a dictionary with the number of characters for each word
# 

list_of_fruits = ["Apple", "Banana", "Orange", "Strawberry"]

word_lengths = { fruit : len(fruit) for fruit in list_of_fruits }

print(word_lengths)
````
