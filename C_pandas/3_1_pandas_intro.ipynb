## What is Pandas and why is it useful?

**Pandas is a Python library designed for data manipulation and analysis. It provides fast, flexible tools for working with structured data.**

Pandas is especially powerful for working with tabular data (data organized in rows and columns).

**Key advantages:**
- Makes data cleaning, transformation, and exploration simple and intuitive
- Integrates seamlessly with other scientific libraries
- Built on top of NumPy, which makes it very fast and efficient for data manipulation

<br>

**Why it matters in AI/ML:**
- Many machine learning projects involve structured tabular data, often in messy real-world form (e.g. missing data, data in the wrong format, etc.)
- Pandas is used to explore, clean, and transform that data
- It is an excelent tool to prepare datasets before they are passed into ML frameworks like scikit-learn or TensorFlow

→ In short: Pandas is a core tool for data cleaning and preprocessing in AI/ML workflows.

<br>

**Why do we need to "clean" our data, you may ask.**

In most Machine Learning projects, we work with real-world data, which often contains missing values, incorrect formats, duplicates, and other inconsistencies that need to be fixed before training a model.

## Pandas vs NumPy

- NumPy → provides efficient numerical arrays and low-level mathematical operations
- Pandas → builds on NumPy and adds high-level data structures for working with labeled, structured data

→ Pandas is essentially a more user-friendly layer on top of NumPy, optimized for data analysis rather than pure computation.

<br>

## Series vs. DataFrame


At its core, Pandas is built around two main data structures:
- Series → a one-dimensional labeled array (like a single column of data)
- DataFrame → a two-dimensional table with rows and columns (similar to an Excel sheet or SQL table)

## Pandas Series

A Series is a 1-dimensional labeled array (like a single column in a table).
- Has values
- Has an index (labels)


````python
import pandas as pd

revenue = pd.Series(
    [1200, 1350, 1500, 1750],
    index=["Jan", "Feb", "Mar", "Apr", ]
    )

print(type(revenue))
print(revenue)
````

````python
print(revenue["Apr"])
print(revenue.mean())
````

<br>

## Pandas DataFrames


- A DataFrame is a 2-dimensional labeled table (rows + columns).
- It’s basically a collection of Series sharing the same index.





````python
# One common way to create a DataFrame is from a dictionary, where:
# - each key represents a column name, 
# - and the associated value is a list or array of data for that column.

customer_data = {
    "customer_id": [1, 2, 3],
    "country": ["Spain", "France", "Germany"],
    "monthly_spend": [49.9, 29.9, 99.9]
}

df = pd.DataFrame(customer_data)

print(type(df))
display(df)
````

````python
print(df["monthly_spend"])
````

````python
print(df["monthly_spend"].mean())
````

## Loading data from different sources

You can create DataFrames from various data sources, including dictionaries, lists, files, and databases.

From Python data structures:
- Dictionary of lists: each key becomes a column, and the lists provide column values.
- List of dictionaries: each dictionary becomes a row, and keys map to column names.

From files:
- Pandas supports many different files, like CSV, Excel (xlsx, xls, xlsm, xlsb), ODF (odf, ods, odt)...
- Example I (CSV): `df = pd.read_csv('file.csv')`
- Example II (Excel): `df = pd.read_excel('file.xlsx')`

From databases
- You can also load data from databases
- Example using SQL queries: `df = pd.read_sql('SELECT * FROM table', engine)`

Info provided by `df.describe()`:

- `count`: The number of non-null (valid) entries in each column.
- `mean`: The average value of the data in the column.
- `std`: The standard deviation, which measures how spread out the values are from the mean.
- `min`: The smallest value present in the column.
- `25% (First Quartile)`: The value below which 25% of the data falls; it marks the lower boundary of the middle 50% of the data.
- `50% (Median)`: The middle value of the dataset, separating the higher half from the lower half.
- `75% (Third Quartile)`: The value below which 75% of the data falls; it marks the upper boundary of the middle 50% of the data.
- `max`: The largest value present in the column.

<br>

Median and Quartiles:
- First Quartile (Q1): 
    - The value at or below which 25% of the data falls.
    - Example: If Q1 = 18 years, then 25% of the people are 18 years old or younger, and 75% are 18 or older.
- Median (Q2): 
    - The value at or below which 50% of the data falls (the middle value).
    - Example: If the median = 30 years, then 50% of the people are 30 years old or younger, and 50% are 30 or older.
- Third Quartile (Q3): 
    - The value at or below which 75% of the data falls.
    - Example: If Q3 = 50 years, then 75% of the people are 50 years old or younger, and 25% are older than 50.

## Order a dataframe

````python
#
# Example I: order by one column
#

new_df = avocado_df[["Date", "AveragePrice", "Total Volume", "type", "year", "region"]]

print("before ordering...")
display(new_df.head(5))

new_df = new_df.sort_values(by="AveragePrice", ascending=True)
print("after ordering...")
display(new_df.head(5))

````

````python
#
# Example II: order by multiple columns
#

new_df = avocado_df[["Date", "AveragePrice", "Total Volume", "type", "year", "region"]]

print("before ordering...")
display(new_df.head(5))

new_df = new_df.sort_values(by=["AveragePrice", "region"], ascending=True)
print("after ordering...")
display(new_df.head(5))


# notes on the order:
# - by=["AveragePrice", "region"] → sorts first by "AveragePrice", then breaks ties using "region".
# - by=["region", "AveragePrice"] → sorts first by "region", then breaks ties using "AveragePrice".

# If you want different orders per column, you can pass a list:
# df.sort_values(by=["AveragePrice", "region"], ascending=[False, True])

````

## NaN values

In pandas, missing values are represented as NaN (Not a Number). 


````python
import numpy as np 

data = {
    "year": [2020, 2021, 2020, 2003, 2007],
    "town": ["Paris", "Lyon", "Marseille", "Paris", "Toulouse"],
    "age": [47, 35, np.nan, np.nan, 72], # note: some ages are missing (we're using np.nan, can also use None)
}

people_df = pd.DataFrame(data)

people_df
````

````python
#
# df.info():
# - provides info on the number of entries, number of columns, Non-Null Count, and data types (dtype).
#

people_df.info()  
````

````python
#
# Check missing values in the DataFrame
# (this will generate a dataframe with True/False values, indicating where the original df has missing values)
#
missing_values = people_df.isna()
display(missing_values)

#
# Count missing values in each colum
# (sums True values per column, True=1, False=0, giving the number of missing values in each column)
#
missing_counts = missing_values.sum()
display(missing_counts)

# Count columns with missing values
number_of_columns_with_missing_values = missing_counts[missing_counts > 0].count()
print(f"Number of columns with missing data: {number_of_columns_with_missing_values}")
````

<br>

Why Pandas represents missing values as NaN (Not a Number) instead of None?

- Pandas is built on NumPy, and relies on fast, vectorized arrays that work best with numeric types. `NaN` is a standard floating-point value, so it can live inside numeric arrays without breaking computations. 
- On the other hand, using `None` would force the data into a slower "object" type, which reduces performance and disables many efficient mathematical operations.
